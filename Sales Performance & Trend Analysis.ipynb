## Duplicates

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
```

## Data Types Correction

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

## Feature Engineering

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.to_period('Q')
```

## Verifying Data

```python
print(df.info())
print(df.isnull().sum())
```

# Exploratory Data Analysis (EDA)


## Basic Sales Insights


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Total sales, profit, orders
print("Total Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())
print("Total Orders:", df['Order ID'].nunique())
```

## Sales Over Time


```python
# Monthly sales trend
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M')).sum(numeric_only=True)['Sales']

monthly_sales.plot(kind='line', figsize=(12,6), marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

## Sales by Product Category


```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

sns.barplot(x=category_sales.values, y=category_sales.index, palette="viridis")
plt.title("Sales by Product Category")
plt.show()
```

## Sales by Customer Segment


```python
segment_sales = df.groupby('Segment')['Sales'].sum()

sns.barplot(x=segment_sales.index, y=segment_sales.values, palette="mako")
plt.title("Sales by Customer Segment")
plt.show()
```

## Sales by Region


```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

sns.barplot(x=region_sales.index, y=region_sales.values, palette="cubehelix")
plt.title("Sales by Region")
plt.show()
```

## Top 10 Products


```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_products.values, y=top_products.index, palette="plasma")
plt.title("Top 10 Products by Sales")
plt.show()
```

# Business Insights & Recommendations


## Sales Trends Over Time


```python
# Monthly Sales
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M')).sum(numeric_only=True)['Sales']

# Display top 5 months by sales
print(monthly_sales.sort_values(ascending=False).head())

# Plot trend
monthly_sales.plot(kind='line', figsize=(12,6), marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.show()
```

## Sales by Product Category


```python
category_sales = df.groupby('Category')[['Sales','Profit']].sum().sort_values('Sales', ascending=False)
print(category_sales)

sns.barplot(x=category_sales['Sales'], y=category_sales.index, palette="viridis")
plt.title("Sales by Product Category")
plt.show()
```

## Sales by Customer Segment


```python
segment_sales = df.groupby('Segment')[['Sales','Profit']].sum()
print(segment_sales)

sns.barplot(x=segment_sales['Sales'], y=segment_sales.index, palette="mako")
plt.title("Sales by Customer Segment")
plt.show()
```

## Sales by Region


```python
region_sales = df.groupby('Region')[['Sales','Profit']].sum().sort_values('Sales', ascending=False)
print(region_sales)

sns.barplot(x=region_sales['Sales'], y=region_sales.index, palette="cubehelix")
plt.title("Sales by Region")
plt.show()
```

## Top Products


```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)

sns.barplot(x=top_products.values, y=top_products.index, palette="plasma")
plt.title("Top 10 Products by Sales")
plt.show()
```

## Profitability Analysis

```python
profitability = df.groupby('Category')[['Sales','Profit']].sum()
print(profitability)

# Profit vs Sales Scatter
sns.scatterplot(x='Sales', y='Profit', hue='Category', data=df)
plt.title("Profit vs Sales by Category")
plt.show()
```
